```python
# Install Graphviz in Google Colab
!apt-get install graphviz -y
!pip install graphviz

from graphviz import Digraph
from IPython.display import Image

# Create the decision tree graph
def plot_decision_tree():
    tree = Digraph("Decision Tree", format="png")

    # Nodes
    tree.node("1", "Age ≤ 30?")
    tree.node("2", "Income ≤ 25k?")
    tree.node("3", "Income ≤ 50k?")
    tree.node("4", "Age ≤ 45?")
    tree.node("5", "R1")
    tree.node("6", "R2")
    tree.node("7", "R3")
    tree.node("8", "R4")
    tree.node("9", "R5")
    tree.node("10", "R6")

    # Edges
    tree.edge("1", "2", "Yes")
    tree.edge("1", "3", "No")
    tree.edge("2", "5", "Yes")
    tree.edge("2", "6", "No")
    tree.edge("3", "4", "Yes")
    tree.edge("3", "9", "No")
    tree.edge("4", "7", "Yes")
    tree.edge("4", "8", "No")
    tree.edge("3", "4")

    # Render and display the decision tree
    tree.render("decision_tree", format="png")
    return Image(filename="decision_tree.png")

# Call the function to plot the decision tree
plot_decision_tree()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Plot the partitioned 2D feature space
def plot_feature_space():
    # Create the plot
    fig, ax = plt.subplots(figsize=(8, 6))

    # Add partitions as rectangles
    ax.add_patch(Rectangle((0, 0), 30, 25, edgecolor="blue", facecolor="none", lw=2, label="R1"))
    ax.add_patch(Rectangle((0, 25), 30, 75, edgecolor="red", facecolor="none", lw=2, label="R2"))
    ax.add_patch(Rectangle((30, 0), 15, 50, edgecolor="green", facecolor="none", lw=2, label="R3"))
    ax.add_patch(Rectangle((30, 50), 15, 50, edgecolor="orange", facecolor="none", lw=2, label="R4"))
    ax.add_patch(Rectangle((45, 0), 15, 50, edgecolor="purple", facecolor="none", lw=2, label="R5"))
    ax.add_patch(Rectangle((45, 50), 15, 50, edgecolor="brown", facecolor="none", lw=2, label="R6"))

    # Add labels
    ax.text(15, 12.5, "R1", fontsize=12, ha="center", va="center")
    ax.text(15, 62.5, "R2", fontsize=12, ha="center", va="center")
    ax.text(37.5, 25, "R3", fontsize=12, ha="center", va="center")
    ax.text(37.5, 75, "R4", fontsize=12, ha="center", va="center")
    ax.text(52.5, 25, "R5", fontsize=12, ha="center", va="center")
    ax.text(52.5, 75, "R6", fontsize=12, ha="center", va="center")

    # Set axis limits and labels
    ax.set_xlim(0, 60)
    ax.set_ylim(0, 100)
    ax.set_xlabel("Age")
    ax.set_ylabel("Income (in $1000s)")
    ax.set_title("2D Feature Space Partitioning")
    ax.legend()
    plt.grid()
    plt.show()

# Call the function to plot
plot_feature_space()
```
